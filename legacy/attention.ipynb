```python
from torch import nn
import torch

class Embedding:
  def __init__(self):
    # 代入する必要あり。
    self.embedding = None # nn.Embedding()を入れる
    self.vocab_size = None
    self.embed_dim = None
    self.ids = None

  def exec(self):
    if self.embedding is None:
      self.embedding = nn.Embedding(self.vocab_size, self.embed_dim)
    ids_tensor = torch.tensor(self.ids)
    tensor = self.embedding(ids_tensor)
    return tensor.detach().numpy()
  
class LinearTransform:
  def qkv(self, x):
    q = self.identical_transform_q(x)
    k = self.identical_transform_k(x)
    v = self.identical_transform_v(x)
    return q, k, v
  
  def identical_transform_k(x):
    # keys = []
    # for i in range(len(x)):
    #   keys.append(x[i])
    # return keys
    return x.copy()
  def identical_transform_v(x):
    # values = []
    # for i in range(len(x)):
    #   values.append(x[i])
    # return values
    return x.copy()
  def identical_transform_q(x):
    # queries = []
    # for i in range(len(x)):
    #   queries.append(x[i])
    # return queries
    return x.copy()
    
```

```python
import numpy as np
from scipy.special import softmax
from torch import nn
import torch

# Attention計算の目標は、「一次元のID配列から、ベクトルの列に変換」すること

def identical_transform_k(x):
  # keys = []
  # for i in range(len(x)):
  #   keys.append(x[i])
  # return keys
  return x.copy()
def identical_transform_v(x):
  # values = []
  # for i in range(len(x)):
  #   values.append(x[i])
  # return values
  return x.copy()
def identical_transform_q(x):
  # queries = []
  # for i in range(len(x)):
  #   queries.append(x[i])
  # return queries
  return x.copy()

def qkv(x):
  q = identical_transform_q(x)
  k = identical_transform_k(x)
  v = identical_transform_v(x)
  return q, k, v

# 一次元（idだから）
ids = np.array([1,3])

embed = Embedding()
embed.vocab_size = 10
embed.embed_dim = 8
embed.ids = ids
embedding = embed.exec()

q, k, v = qkv(embedding)

# qとkの内積を計算
dots = []
for i in range(len(q)):
  row = []
  for j in range(len(k)):
    dot = np.dot(q[i], k[j])
    row.append(dot)
  dots.append(row)

print(dots)

# 重みは基準語ごとにつくる
weights = []
for i in range(len(dots)):
  weights.append(softmax(dots[i]))

# np配列にする。
weights = np.array(weights)

# 重みつき平均をだして、新しい表現
ys = []
for i in range(len(weights)):
  y = np.zeros_like(v[0], float)
  for j in range(len(weights[i])):
    y += weights[i][j] * v[j]
  ys.append(y)

ys

# ここから先はfeed-forward

  
```
